get the category ids of the 10 longest movies (descending) released after 2005

```python
!psql -c "SELECT pagila.film.title, pagila.film_category.category_id FROM pagila.film INNER JOIN pagila.film_category ON pagila.film_category.film_id = pagila.film.film_id WHERE pagila.film.release_year > 2005 ORDER BY pagila.film.length DESC LIMIT 10"
```

Get the names and addresses of the active customers ordered by id

```python
!psql -c "SELECT pagila.customer.customer_id, pagila.customer.first_name, pagila.customer.last_name, pagila.address.address FROM pagila.address INNER JOIN pagila.customer ON pagila.customer.address_id = pagila.address.address_id ORDER BY pagila.customer.customer_id ASC LIMIT 10"
```

Get staff names alphabetically for purchases over $10

```python
!psql -c "SELECT pagila.staff.last_name, pagila.staff.first_name, pagila.payment.amount FROM pagila.payment INNER JOIN pagila.staff ON pagila.staff.staff_id = pagila.payment.staff_id WHERE pagila.payment.amount > 10 ORDER BY pagila.staff.last_name DESC LIMIT 10"
```

Get country, city names for cities that are not in algeria sorted by country name alphabetically

```python
!psql -c "SELECT pagila.country.country, pagila.city.city FROM pagila.city FULL OUTER JOIN pagila.country ON pagila.city.country_id = pagila.country.country_id WHERE NOT pagila.country.country='Algeria' ORDER BY pagila.country.country ASC LIMIT 10"
```

Get all inventory ids for every item with id less than 10, ordered by store id and inventory id

```python
!psql -c "SELECT pagila.store.store_id, pagila.inventory.inventory_id FROM pagila.inventory FULL OUTER JOIN pagila.store ON pagila.store.store_id = pagila.inventory.store_id WHERE pagila.inventory.inventory_id < 10 ORDER BY pagila.store.store_id ASC, pagila.inventory.inventory_id ASC LIMIT 10"
```

combine address and city into one table filling the blanks with null values for adddress ids less than 50, reverse alphabetical

```python
!psql -c "SELECT * FROM pagila.address FULL OUTER JOIN pagila.city ON pagila.city.city_id = pagila.address.city_id WHERE pagila.address.address_id < 50 ORDER BY pagila.city.city DESC"
```

Create all 1on1 pairings of customers, ordered by first customer last name

```python
!psql -c "SELECT A.last_name AS customerA_last_name, A.first_name AS customerA_first_name, B.last_name AS customerB_last_name, B.first_name AS customerB_first_name FROM pagila.customer A, pagila.customer B WHERE NOT ((A.first_name = B.first_name) OR (A.last_name = B.last_name)) ORDER BY customerA_last_name ASC LIMIT 10"
```
