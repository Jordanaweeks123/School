```python
import pandas as pd
from google.cloud import firestore
db = firestore.Client()
batch = db.batch()

df = pd.read_csv('shopify/apps.csv', sep=',', header=0, lineterminator='\n')
apps = df.values.tolist()

df = pd.read_csv('shopify/key_benefits.csv', sep=',', header=0, lineterminator='\n')
kbs = df.values.tolist()

df = pd.read_csv('shopify/pricing_plans.csv', sep=',', header=0, lineterminator='\n')
pps = df.values.tolist()

df = pd.read_csv('shopify/pricing_plan_features.csv', sep=',', header=0, lineterminator='\n')
ppfs = df.values.tolist()

counter = 0

for row in apps:
    record = {}
    record['id'] = row[0]
    #record['url'] = row[1]
    record['title'] = row[2]
    record['developer'] = row[3]
    #record['developer_link'] = row[4]
    #record['icon'] = row[5]
    record['rating'] = row[6]
    record['reviews_count'] = row[7]
    
    apps_ref = db.collection('apps').document(row[0])
        
    batch.set(apps_ref, record)
    counter += 1
    
    for kb in kbs:
        if(row[0] == kb[0]):
            record = {}
            record['title'] = kb[1]
            record['description'] = kb[2]

            kb_ref = apps_ref.collection('key_benefits').document(kb[0])

            batch.set(kb_ref, record)  
            
            counter += 1
            if counter >= 499:
                batch.commit()
                counter = 0
    for pp in pps:
        if(row[0] == pp[1]):
            record = {}
            record['id'] = pp[0]
            record['title'] = pp[2]
            record['price'] = pp[2]
            for ppf in ppfs:
                if ppf[1] == pp[0]:
                    record['feature'] = ppf[2]
                    

            pp_ref = apps_ref.collection('pricing_plans').document(pp[1])

            batch.set(kb_ref, record)  
            
            counter += 1
            if counter >= 499:
                batch.commit()
                counter = 0
                
    batch.commit()
```

```python
import pandas as pd
from google.cloud import firestore
db = firestore.Client()
batch = db.batch()

df = pd.read_csv('shopify/categories.csv', sep=',', header=0, lineterminator='\n')
categories = df.values.tolist()

df = pd.read_csv('shopify/apps.csv', sep=',', header=0, lineterminator='\n')
apps = df.values.tolist()

df = pd.read_csv('shopify/apps_categories.csv', sep=',', header=0, lineterminator='\n')
acs = df.values.tolist()

counter = 0

for row in categories:
    record = {}
    record['id'] = row[0]
    record['title'] = row[1]
    
    categories_ref = db.collection('categories').document(row[0])
        
    batch.set(categories_ref, record)
    counter += 1
    
    for ac in acs:
        if(ac[1] == row[0]):
            record = {}
            for app in apps:
                if ac[0] == app[0]:
                    record['id'] = app[0]
                    record['title'] = app[2]
                    record['developer'] = app[3]
                    record['rating'] = app[6]
                    record['reviews_count'] = app[7]

                    apps_ref = categories_ref.collection('apps').document(app[0])

                    batch.set(apps_ref, record)  
            
                    counter += 1
                    if counter >= 499:
                        batch.commit()
                        counter = 0
                
    batch.commit()
```

```python
import pandas as pd
from google.cloud import firestore
db = firestore.Client()
batch = db.batch()

queries = db.collection('categories').stream()

print("Subcollections per Category")

for query in queries:
    counter = 0
    collections = db.collection('categories').document(query.id).collections()
    for collection in collections:
        for doc in collection.stream():
            counter += 1
    print(query.get('title'), counter)
```

```python
import pandas as pd
from google.cloud import firestore
db = firestore.Client()
batch = db.batch()

queries = db.collection('apps').stream()

print("Subcollections per App (Limit 20)")

limiter = 0

for query in queries:
    counter = 0
    collections = db.collection('apps').document(query.id).collections()
    for collection in collections:
        for doc in collection.stream():
            counter += 1
    print(query.get('title'), counter)
    limiter += 1
    if limiter >= 20:
        break
```

```python
import pandas as pd
from google.cloud import firestore
db = firestore.Client()
batch = db.batch()

queries = db.collection('categories').where('title','==','Productivity').stream()

print("Top 10 Productivity Apps")

apps = []
ratings = []
for query in queries:
    collections = db.collection('categories').document(query.id).collections()
    for collection in collections:
        for doc in collection.stream():
            app = doc.to_dict()
            apps.append(app)

print(sorted(apps, key = lambda i: i['ratings'],reverse=True))
```
